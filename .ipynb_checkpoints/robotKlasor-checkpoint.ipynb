```python
import openpyxl

def main():
    # Döküman dosyasının adı
    dokuman_adi = "dokuman.txt"
    
    # Excel dosyasının adı
    excel_dosyasi_adi = "hata_mesajlari.xlsx"

    # Excel dosyasını oluştur
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Hata Mesajları"

    # Dökümandan ER0 ile başlayan satırları oku ve Excel dosyasına yaz
    with open(dokuman_adi, "r", encoding="utf-8") as file:
        satirlar = file.readlines()
        hata_mesajlari = [satir.strip() for satir in satirlar if satir.startswith("ER0")]

        for index, hata_mesaji in enumerate(hata_mesajlari):
            sheet.cell(row=index+1, column=1, value=hata_mesaji)

    # Excel dosyasını kaydet
    workbook.save(excel_dosyasi_adi)
    print("Hata mesajları başarıyla Excel dosyasına yazıldı.")

if __name__ == "__main__":
    main()
```

```python
import openpyxl

def main():
    # Excel dosyasının adı
    excel_dosyasi_adi = "hexadecimal.xlsx"

    # Excel dosyasını oluştur
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Hexadecimal"

    # 1'den 566'ya kadar olan sayıları hexadecimal olarak yaz
    for i in range(1, 567):
        hexadecimal_deger = hex(i)[2:].upper()  # 0x kısmını almaz ve harfleri büyük harf olarak yazar
        sheet.cell(row=i, column=1, value=hexadecimal_deger)

    # Excel dosyasını kaydet
    workbook.save(excel_dosyasi_adi)
    print("1'den 566'ya kadar olan sayılar hexadecimal olarak Excel dosyasına yazıldı.")

if __name__ == "__main__":
    main()
```

```python
import openpyxl

def main():
    # Döküman dosyasının adı
    dokuman_adi = "dokuman.txt"
    
    # Excel dosyasının adı
    excel_dosyasi_adi = "hata_mesajlari.xlsx"

    # Excel dosyasını oluştur
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Hata Mesajları"

    # Dökümandan ER0 ile başlayan ve "=" sembolüne kadar olan satırları oku ve Excel dosyasına yaz
    with open(dokuman_adi, "r", encoding="utf-8") as file:
        satirlar = file.readlines()
        hata_mesajlari = [satir.split("=")[0].strip() for satir in satirlar if satir.startswith("ER0")]

        for index, hata_mesaji in enumerate(hata_mesajlari):
            sheet.cell(row=index+1, column=1, value=hata_mesaji)

    # Excel dosyasını kaydet
    workbook.save(excel_dosyasi_adi)
    print("Hata mesajları başarıyla Excel dosyasına yazıldı.")

if __name__ == "__main__":
    main()
```

```python
import openpyxl

def hex_to_decimal(hex_str):
    # hexadecimal sayıyı decimal sayıya çevirir
    return int(hex_str, 16)

def main():
    # Döküman dosyasının adı
    dokuman_adi = "dokuman.txt"
    
    # Excel dosyasının adı
    excel_dosyasi_adi = "hata_mesajlari.xlsx"

    # Excel dosyasını oluştur
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Hata Mesajları"

    # Dökümandan ER0 ile başlayan ve hexadecimal sayıları içeren satırları oku ve Excel dosyasına yaz
    with open(dokuman_adi, "r", encoding="utf-8") as file:
        satirlar = file.readlines()
        satir_numarasi = 1
        for satir in satirlar:
            if satir.startswith("ER0"):
                hex_sayilar = satir.strip().split()[1:]  # ER0 ile başlayan satırdaki hexadecimal sayıları al
                for hex_sayi in hex_sayilar:
                    # İlk sütuna hexadecimal sayıyı yaz
                    sheet.cell(row=satir_numarasi, column=1, value=hex_sayi)
                    # İkinci sütuna hexadecimal sayının karşılık geldiği decimal sayıyı yaz
                    sheet.cell(row=satir_numarasi, column=2, value=hex_to_decimal(hex_sayi))
                    satir_numarasi += 1

    # Excel dosyasını kaydet
    workbook.save(excel_dosyasi_adi)
    print("Hata mesajları ve hexadecimal sayılar Excel dosyasına yazıldı.")

if __name__ == "__main__":
    main()
```

```python
import openpyxl

def main():
    # Döküman dosyasının adı
    dokuman_adi = "dokuman.txt"
    
    # Excel dosyasının adı
    excel_dosyasi_adi = "hata_mesajlari.xlsx"

    # Excel dosyasını oluştur
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Hata Mesajları"

    # Dökümandan ER0 ile başlayan satırları oku ve Excel dosyasına yaz
    with open(dokuman_adi, "r") as file:
        satirlar = file.readlines()
        hata_mesajlari = [satir.replace("ER0", "").strip() for satir in satirlar if satir.startswith("ER0")]

        for index, hata_mesaji in enumerate(hata_mesajlari):
            sheet.cell(row=index+1, column=1, value=hata_mesaji)

    # Excel dosyasını kaydet
    workbook.save(excel_dosyasi_adi)
    print("Hata mesajları başarıyla Excel dosyasına yazıldı.")

if __name__ == "__main__":
    main()
```

```python
import openpyxl

def hex_to_decimal(hex_str):
    # hexadecimal sayıyı decimal sayıya çevirir
    return int(hex_str, 16)

def main():
    # Döküman dosyasının adı
    dokuman_adi = "dokuman.txt"
    
    # Excel dosyasının adı
    excel_dosyasi_adi = "decimal_sayilar.xlsx"

    # Excel dosyasını oluştur
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Decimal Sayılar"

    # Dökümandan hexadecimal sayıları oku ve Excel dosyasına yaz
    with open(dokuman_adi, "r", encoding="utf-8") as file:
        satirlar = file.readlines()
        satir_numarasi = 1
        for satir in satirlar:
            hex_sayilar = satir.strip().split()  # Satırdaki hexadecimal sayıları al
            for hex_sayi in hex_sayilar:
                # İlk sütuna hexadecimal sayıyı yaz
                sheet.cell(row=satir_numarasi, column=1, value=hex_sayi)
                # İkinci sütuna hexadecimal sayının karşılık geldiği decimal sayıyı yaz
                sheet.cell(row=satir_numarasi, column=2, value=hex_to_decimal(hex_sayi))
                satir_numarasi += 1

    # Excel dosyasını kaydet
    workbook.save(excel_dosyasi_adi)
    print("Hexadecimal sayılar decimal olarak Excel dosyasına yazıldı.")

if __name__ == "__main__":
    main()
```

```python
import openpyxl

def main():
    # Döküman dosyasının adı
    dokuman_adi = "dokuman.txt"
    
    # Excel dosyasının adı
    excel_dosyasi_adi = "sayilar.xlsx"

    # Excel dosyasını oluştur
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Sayılar"

    # Dökümandan sayıları oku ve Excel dosyasına yaz
    with open(dokuman_adi, "r") as file:
        satirlar = file.readlines()
        satir_numarasi = 1
        for satir in satirlar:
            sayi = satir.strip()
            if sayi.startswith("0"):  # Sayının başında 0 varsa sil
                sayi = sayi.lstrip("0")
            sheet.cell(row=satir_numarasi, column=1, value=sayi)
            satir_numarasi += 1

    # Excel dosyasını kaydet
    workbook.save(excel_dosyasi_adi)
    print("Sayılar Excel dosyasına yazıldı.")

if __name__ == "__main__":
    main()
```
