```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# kodun çıktılarını doğrudan hücre içinde görmek için kullanılır.
# plt.show() kullanmanıza gerek kalmaz
```

```python
yasListe = [10,20,30,30,30,40,50,60,70,75]
kiloListe = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListe)
numpyKiloListesi = np.array(kiloListe)
# listeleri numpy formatına çevirelim. Uyum ve kolaylık için.
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
# sırayla satır ve sütun olarak bastı. renk de green
```

```python
plt.plot(numpyKiloListesi,numpyYasListesi,"g")
# eksenleri değiştirdik ama bu çok anlamlı bir grafik olmuyor
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")

plt.xlabel("Yas")
plt.ylabel("Kilo")
# plt.show() # %matplotlib inline  kullanıldığı için gerek kalmıyor
plt.title("Kilonun Yasa Gore Degisimi")
# eksenlere isim verebiliriz
```

```python
numpyDizisi1 = np.linspace(0,10,20)
# 0 ile 10 arası 20 sayı basar. Eşit aralıklarla.
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
# "g--" sayesinde -- görünümlü çizgi
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
# "b*-" sayesinde *- görünümlü çizgi
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-") # matrisin [1 1] kısmı

plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g--") # matrisin [1 2] kısmı

# plt.subplot(kaç sıra olacak, kaç grafik gösterilecek, hangisi çizilecek)
```

```python
benimFigure = plt.figure()
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri Ismi")
figureAxes.set_ylabel("Y Ekseni Veri Ismi")
figureAxes.set_title("Grafik Başligi")
```

```python
benimFigure = plt.figure()

figureAxes = benimFigure.add_axes([0.1, 0.1, 0.5, 0.5])
# add_axes([left, bottom, width, height])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")

figureAxes.set_xlabel("X Ekseni Veri Ismi")
figureAxes.set_ylabel("Y Ekseni Veri Ismi")

figureAxes.set_title("Grafik Başligi")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.25,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başligi")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başligi")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout()
# grafik mesafelerini düzenler
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend()
# Eksenin uygun bir yerine bu açıklama kutusunu ekler.
# Kutunun konumu varsayılan olarak en iyi görünen yere (örneğin sağ üst köşe) yerleştirilir.
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=1,fontsize=10,frameon=False,ncol=2)
# farklı parametreler kullanılabilir
```

```python
yeniFigur = plt.figure(figsize=(3,3))
# figsize ile de width ve height ayarlanabilir.

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=1,fontsize=10,frameon=False,ncol=2)
```

```python
yeniFigur = plt.figure(dpi=150)
# "Dots Per Inch" (İnç başına düşen nokta sayısı). Çözünürlük ayarlar.

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=1,fontsize=10,frameon=False,ncol=2)
```
